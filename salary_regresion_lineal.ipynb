## Pre Procesado

```python
#Importar librerias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importar el dataset
dataset = pd.read_csv("Salary_Data.csv")
x = dataset.iloc[:,0].values
y = dataset.iloc[:,1].values
```

```python
#Dividir el datased en conjunto de entrenamiento y testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=1/3, random_state=0)
```

## Creacin del modelo

```python
#Crear modelo de Regresion Lineal Simple con conjunto de entranamiento
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train.reshape(-1, 1), y_train.reshape(-1, 1))
```

## Prediccin

```python
#Predecir el conjunto de test
y_pred = regression.predict(x_test.reshape(-1, 1))
```

## Visualizacin de datos

```python
#Visualizar los resultados de entrenamiento
plt.scatter(x_train,y_train, color="red")
plt.plot(x_train, regression.predict(x_train.reshape(-1, 1)),color="blue")
plt.title("Sueldo VS aos de experiencia (Conjunto de entrenamiento)")
plt.xlabel("Aos de experiencia")
plt.ylabel("Sueldo en $")
plt.show()
```

```python
#Visualizar los resultados de test
plt.scatter(x_test,y_test, color="red")
plt.plot(x_test, y_pred,color="blue")
plt.title("Sueldo VS aos de experiencia (Conjunto de testing)")
plt.xlabel("Aos de experiencia")
plt.ylabel("Sueldo en $")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# Veamos los coeficienetes obtenidos, En nuestro caso, sern la Tangente
print('Coefficients: \n', regression.coef_)
# Este es el valor donde corta el eje Y (en X=0)
print('Independent term: \n', regression.intercept_)
# Error Cuadrado Medio
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
# Puntaje de Varianza. El mejor puntaje es un 1.0
print('Variance score: %.2f' % r2_score(y_test, y_pred))
```
