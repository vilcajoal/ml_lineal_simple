```python
#Importar librerias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importar el dataset
dataset = pd.read_csv("articulos_ml.csv")
x = dataset.iloc[:,2].values
y = dataset.iloc[:,7].values
```

```python
#Dividir el datased en conjunto de entrenamiento y testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=1/4, random_state=0)
```

```python
#Creación del modelo de Regresión Lineal Simple con conjunto de entrenamiento
from sklearn.linear_model import LinearRegression
regre = LinearRegression()
regre.fit(x_train.reshape(-1, 1), y_train.reshape(-1, 1))
x_test.shape
x_train.shape
```

```python
#Predecir el conjunto de test
#y_pred = regression.predict(x_test.reshape(-1, 1))
y_predi = regre.predict(x_test.reshape(-1, 1))
```

```python
#Visualizar los resultados de entrenamiento
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train, regre.predict(x_train.reshape(-1,1)),color="blue") 
plt.title("Número de palabras VS Número de compartidas")
plt.xlabel("Número de palabras")
plt.ylabel("Número de compartidas")
plt.show()
```

```python
#Visualizar los resultados de test
plt.scatter(x_test,y_test, color="red")
plt.plot(x_test, y_predi,color="blue")
plt.title("Número de palabras VS Número de compartidas")
plt.xlabel("Número de palabras")
plt.ylabel("Número de compartidas")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# Veamos los coeficienetes obtenidos, En nuestro caso, serán la Tangente
print('Coefficients: \n', regre.coef_)
# Este es el valor donde corta el eje Y (en X=0)
print('Independent term: \n', regre.intercept_)
# Error Cuadrado Medio
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_predi))
# Puntaje de Varianza. El mejor puntaje es un 1.0
print('Variance score: %.2f' % r2_score(y_test, y_predi))
```
